```python
import numpy as np
import pandas as pd
from collections import Counter
```

```python
plt.style.use("fivethirtyeight")
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head(5)
```

```python
ids = df['Responder_id']
lang_responses = df['LanguagesWorkedWith']
```

```python
# Removing the semi colons
```

```python
language_counter = Counter()

for response in lang_responses:
    language_counter.update(response.split(';'))
```

```python
languages = []
popularity = []
for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])
```

```python
plt.barh(languages, popularity)
```

```python
languages.reverse()
popularity.reverse()
```

```python
plt.barh(languages, popularity)
```

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.1, 0]
```

```python
plt.pie(slices, labels=labels, explode=explode, shadow=True,
        startangle=90, autopct='%1.1f%%',
        wedgeprops={'edgecolor': 'black'})

plt.tight_layout()
plt.title("Pie Chart")
```

```python
from itertools import count
from matplotlib.animation import FuncAnimation
# For those using Jupyter Notebook, adding this line at the top of the notebook will make the graph appear: 
%matplotlib notebook
```

```python
x_vals = []
y_vals = []

index = count()

def animate(i):
    x_vals.append(next(index))
    y_vals.append(random.randint(0, 5))
    
    plt.plot(x_vals, y_vals)

# One thousand coz data updates once in 1 sec.
ani = FuncAnimation(plt.gcf(), animate, interval = 1000)
plt.tight_layout()

# In onvscode file
```
